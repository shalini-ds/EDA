###### So we have Gender, Type of Travel, Customer Type and Satisfaction all have binary options. While Class has a 3 different values that populate the column. To make the data regression ready we will translate all the binary options into 1's and 0's, or classically referred to as dummy variables. 

```python
data['Gender'].replace(['Male','Female'],[0,1], inplace = True)
data['Customer Type'].replace(['First-time','Returning'],[0,1], inplace = True)
data['Type of Travel'].replace(['Personal','Business'],[0,1], inplace = True)
data['Class'].replace(['Economy','Economy Plus','Business'],[1,2,3], inplace = True)
data['Satisfaction'].replace(['Neutral or Dissatisfied','Satisfied'],[0,1], inplace = True)
```

# Model Building 

```python
# Now i will seperate my dependent and independent variable from the dataset
x = data.drop('Satisfaction', axis = 'columns')
```

```python
x.head(2)
```

```python
# Now i will assign new varible name to my dependent variable to feed data to model
y = data.Satisfaction

# Lets see dataset
y.head()
```

#### Splitting the data into train and test 

```python
from sklearn.model_selection import train_test_split 
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 10)
```

```python
#Using logistic rgression model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lor = LogisticRegression()
lor.fit(x_train, y_train)
```

```python
lor_pred = lor.predict(x_test)
```

```python
print("Training Accuracy:", lor.score(x_train,y_train))
print("Testing Accuracy:",lor.score(x_test,y_test))
```

```python
print("Logistic Regression Acuuracy Score is:", accuracy_score(lor_pred, y_test))
```

```python
# Using Decision tree model

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
dt_pred = dt.predict(x_test)
```

```python
print("Training Accuracy:", dt.score(x_train,y_train))
print("Testing Accuracy:",dt.score(x_test,y_test))
```

```python
print("DecisionTree Acuuracy Score is:", accuracy_score(dt_pred, y_test))
```

```python
# Using Rnadomforestclassifier

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier() 
```

```python
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
```

```python
print("Training Accuract:", rf.score(x_train, y_train))
print("Testing Accuract:",rf.score(x_test, y_test))
```

```python
print("RandomForest Classifier Acuuracy Score is:", rf.score(x_test, y_test))
```
